```python
import os
import shutil
import numpy as np
from sklearn.model_selection import train_test_split

from google.colab import drive
drive.mount('/content/drive')
```

```python
BASE_DIR = '/content/drive/MyDrive/Machine Learning I/UAS/ML & DL/dataset/'  # Ubah sesuai path folder dataset Anda
CLASSES = ['readable', 'unreadable']

def get_all_files_and_labels(base_dir, classes):
    """Fungsi untuk mengumpulkan semua path file gambar dan labelnya"""
    file_paths = []
    labels = []
    for class_idx, class_name in enumerate(classes):
        class_dir = os.path.join(base_dir, class_name)
        if not os.path.exists(class_dir):
            print(f"Peringatan: Direktori {class_dir} tidak ditemukan.")
            continue
        for filename in os.listdir(class_dir):
            if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
                file_paths.append(os.path.join(class_dir, filename))
                labels.append(class_idx)
    return np.array(file_paths), np.array(labels)
```

```python
# ambil seluruh data
X, y = get_all_files_and_labels(BASE_DIR, CLASSES)
print(f"Total seluruh sampel gambar: {len(X)}")
print(f"Distribusi kelas: {np.bincount(y)}")
```

## Unified Split dan Stratified Split

```python
# subset data yang sama persis untuk ketiga model
X_train_uni, X_temp_uni, y_train_uni, y_temp_uni = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)
X_val_uni, X_test_uni, y_val_uni, y_test_uni = train_test_split(
    X_temp_uni, y_temp_uni, test_size=0.50, random_state=42, stratify=y_temp_uni
)

print(f"[Unified] Semua Model akan menggunakan Test Set yang sama sebesar: {len(X_test_uni)} sampel.")
```

buat struktur file di drive

```python
def create_split_folders(dest_root, X_set, y_set, set_name, classes):
    """Fungsi untuk menyalin file gambar ke folder terpisah (Train/Val/Test)"""
    for class_name in classes:
        os.makedirs(os.path.join(dest_root, set_name, class_name), exist_ok=True)

    for path, label in zip(X_set, y_set):
        class_name = classes[label]
        filename = os.path.basename(path)
        dest_path = os.path.join(dest_root, set_name, class_name, filename)
        shutil.copy(path, dest_path)
```

```python
OUTPUT_DIR = '/content/drive/MyDrive/Machine Learning I/UAS/ML & DL/dataset/split_dataset'

create_split_folders(OUTPUT_DIR, X_train_uni, y_train_uni, 'train', CLASSES)
create_split_folders(OUTPUT_DIR, X_val_uni, y_val_uni, 'val', CLASSES)
create_split_folders(OUTPUT_DIR, X_test_uni, y_test_uni, 'test', CLASSES)
print("\nFolder dataset fisik berhasil dibuat di:", OUTPUT_DIR)
```
